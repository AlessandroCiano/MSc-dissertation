```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file
file_paths = ['./NO-fold/Google/non-thinking/validation/gemini-2.0-flash-lite_calculators_metrics.csv',
              './NO-fold/Google/non-thinking/validation/gemini-2.0-flash_calculators_metrics.csv',
              './NO-fold/Google/non-thinking/validation/gemini-2.5-flash-preview-05-20_calculators_metrics.csv',
              './NO-fold/Google/non-thinking/validation/gemini-2.5-flash-preview-05-20_calculators_ev(noex)_gemini_1_metrics-results.csv',
              './NO-fold/Google/non-thinking/validation/gemini-2.5-pro-preview-05-06_calculators_metrics.csv',
              './NO-fold/DeepSeek/non-thinking/validation/deepseek-chat_calculators_metrics.csv',
              './NO-fold/DeepSeek/non-thinking/validation/deepseek-reasoner_calculators_metrics.csv',
              './NO-fold/Grok/non-thinking/validation/grok-3-latest_calculators_metrics.csv',
              './NO-fold/Grok/non-thinking/validation/grok-3-mini-latest_calculators_metrics.csv']
              

model_names = []
f1_scores = []

for file_path in file_paths:
    # Read the CSV file
    df = pd.read_csv(file_path)

    model_name = file_path.split('/')[-1].split('_')[0]
    
    f1_weight = df['f1_weighted']
    
    model_names.append(model_name)
    f1_scores.append(f1_weight.mean())
    
    print(f"F1 weighted for {model_name}: {f1_weight.mean()}")

model_names = [name.replace('grok', 'Grok').replace('deepseek', 'DeepSeek').replace('gemini', 'Gemini') for name in model_names]
model_names[3] = model_names[3] + '\ncalculators+ (Gemini, noex) prompt'

# Define colors for all models - grouped by provider
colors = ['#4285F4', '#4285F4', '#4285F4', "#003C9C", '#4285F4', # Google models (blues)
          '#ff6b35', '#ff6b35',              # DeepSeek models (oranges)
          '#9c27b0', '#9c27b0']              # Grok models (purples)

# Create bar chart with larger figure size
plt.figure(figsize=(14, 8))
bars = plt.bar(model_names, f1_scores, color=colors)
plt.title('F1 Weighted Scores by Model (non-thinking)', fontsize=16, fontweight='bold')
plt.xlabel('Model', fontsize=12)
plt.ylabel('F1 Weighted Score', fontsize=12)
plt.ylim(0, 1)

# Add value labels on top of bars
for i, v in enumerate(f1_scores):
    plt.text(i, v + 0.01, f'{v:.3f}', ha='center', va='bottom', fontweight='bold', fontsize=10)

plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.grid(axis='y', alpha=0.3)
plt.savefig("non-thinking_scores.pdf") # Save the plot as a PDF
plt.show()
```
