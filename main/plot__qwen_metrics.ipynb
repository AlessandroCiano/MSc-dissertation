```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Load the data
metrics_df = pd.read_csv('/home/s2659893/dissertation/main/Qwen3-0.6B_metrics.csv')

# Display the first few rows
metrics_df.head()
```

```python
# Set the style
plt.figure(figsize=(12, 6))

# Plot loss
#plt.plot(metrics_df['epoch'], metrics_df['eval_loss'], 'b-', marker='o', markersize=4, label='Validation Loss')
plt.plot(metrics_df['epoch'], metrics_df['val_f1_macro'], 'r-', marker='o', markersize=4, label='Validation F1 Macro')
plt.plot(metrics_df['epoch'], metrics_df['val_accuracy'], 'g-', marker='o', markersize=4, label='Validation Accuracy')


plt.title('Loss Values Over Training Epochs', fontsize=16)
plt.xlabel('Epoch', fontsize=14)
plt.ylabel('Loss', fontsize=14)
plt.legend(fontsize=12)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Set the style
plt.figure(figsize=(12, 6))

# Plot loss
plt.plot(metrics_df['epoch'], metrics_df['train_loss'], 'r-', marker='o', markersize=4, label='Training Loss')
plt.plot(metrics_df['epoch'], metrics_df['val_loss'], 'g-', marker='o', markersize=4, label='Validation Loss')


plt.title('Loss Values Over Training Epochs', fontsize=16)
plt.xlabel('Epoch', fontsize=14)
plt.ylabel('Loss', fontsize=14)
plt.legend(fontsize=12)
plt.grid(True)
plt.tight_layout()
plt.show()
```
