```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Load the data
warmup320_df = pd.read_csv('warmup320/metrics/roberta-large_metrics.csv')

# Set the style
fig, ax1 = plt.subplots(figsize=(12, 6))

# Plot F1 weighted and Accuracy on the left y-axis
color = 'tab:red'
ax1.set_xlabel('Epoch', fontsize=14)
ax1.set_ylabel('F1 score', color=color, fontsize=14)
ax1.plot(warmup320_df['epoch'], warmup320_df['eval_f1_weighted'], color=color, linestyle='-', marker='.', markersize=4, label='Validation F1 weighted')
#ax1.plot(warmup320_df['epoch'], warmup320_df['eval_accuracy'], color='green', linestyle='-', marker='s', markersize=4, label='Validation Accuracy')
ax1.tick_params(axis='y', labelcolor=color)
ax1.grid(True)

# Create a second y-axis for the loss
ax2 = ax1.twinx()  # instantiate a second axes that shares the same x-axis

color = 'tab:blue'
ax2.set_ylabel('Loss', color=color, fontsize=14)  # we already handled the x-label with ax1
ax2.plot(warmup320_df['epoch'], warmup320_df['eval_loss'], color=color, linestyle='-', marker='^', markersize=4, label='Validation Loss')
ax2.plot(warmup320_df['epoch'], warmup320_df['train_loss'], color="#003C9C", linestyle='-', marker='^', markersize=4, label='Training Loss')
ax2.tick_params(axis='y', labelcolor=color)

# Add a title
plt.title('RoBERTa training (1271 examples): cross-entropy loss', fontsize=16)

# Add legends
# To combine legends from both axes, we get handles and labels from both and pass them to a single legend command
lines, labels = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax2.legend(lines + lines2, labels + labels2, loc='lower right', fontsize=12)

fig.tight_layout()  # otherwise the right y-label is slightly clipped
plt.savefig("warmup320_FIRST_1fold.pdf") # Save the plot as a PDF
plt.show()
```

```python
# Calculate mean and standard deviation of eval_f1_weighted after epoch 25
filtered_df = warmup320_df[warmup320_df['epoch'] > 25]
mean_f1 = filtered_df['eval_f1_weighted'].mean()
std_f1 = filtered_df['eval_f1_weighted'].std()

print(f"For warmup320_df after epoch 25:")
print(f"Mean F1 weighted score: {mean_f1:.6f}")
print(f"Standard deviation of F1 weighted score: {std_f1:.6f}")

# Let's also calculate the same for other datasets for comparison
datasets = {
    'original': original_df,
    'warmup320': warmup320_df,
    'warmup100': warmup100_df,
    'HALF_warmup200': HALF_warmup200_df,
    'lr1e4_lorar32_alpha64': lr1e4_lorar32_alpha64_df,
    'lr3e5_batch_size4': lr3e5_batch_size4_df
}

print("\nComparison across all datasets after epoch 25:\n")
for name, df in datasets.items():
    filtered = df[df['epoch'] > 25]
    mean = filtered['eval_f1_weighted'].mean()
    std = filtered['eval_f1_weighted'].std()
    max_f1 = filtered['eval_f1_weighted'].max()
    print(f"{name}: Mean={mean:.6f}, Std={std:.6f}, Max={max_f1:.6f}")
```

```python
# Compute mean F1 scores in blocks of 10 epochs for warmup320, original, and warmup100
def mean_f1_by_blocks(df, epoch_col, f1_col, block_size=10):
    df_sorted = df.sort_values(epoch_col)
    blocks = (df_sorted[epoch_col] // block_size).astype(int)
    return df_sorted.groupby(blocks)[f1_col].mean()

# Calculate mean F1 by blocks for each dataset
warmup320_block_means = mean_f1_by_blocks(warmup320_df, 'epoch', 'eval_f1_weighted', block_size=10)
filtered_original = original_df[original_df['epoch'] < 99]
filtered_original_means = mean_f1_by_blocks(filtered_original, 'epoch', 'eval_f1_weighted', block_size=10)
warmup100_block_means = mean_f1_by_blocks(warmup100_df, 'epoch', 'eval_f1_weighted', block_size=10)
HALF_warmup200_block_means = mean_f1_by_blocks(HALF_warmup200_df, 'epoch', 'eval_f1_weighted', block_size=10)

# Align all series to the same index (union of all)
all_blocks = sorted(set(warmup320_block_means.index)
                   .union(set(filtered_original_means.index))
                   .union(set(warmup100_block_means.index))
                   .union(set(HALF_warmup200_block_means.index)))

warmup320_aligned = warmup320_block_means.reindex(all_blocks)
original_aligned = filtered_original_means.reindex(all_blocks)
warmup100_aligned = warmup100_block_means.reindex(all_blocks)
HALF_warmup200_aligned = HALF_warmup200_block_means.reindex(all_blocks)

x = np.arange(len(all_blocks))
width = 0.2  # Narrower width to fit four bars

plt.figure(figsize=(14, 8))
# Plot bars for each dataset with slight offsets
plt.bar(x - 1.5*width, original_aligned, width, label='original', color='tab:orange')
plt.bar(x - 0.5*width, warmup320_aligned, width, label='warmup320', color='tab:blue')
plt.bar(x + 0.5*width, warmup100_aligned, width, label='warmup100', color='tab:green')
plt.bar(x + 1.5*width, HALF_warmup200_aligned, width, label='HALF_warmup200', color='tab:red')

plt.xlabel('Epoch Blocks', fontsize=14)
plt.ylabel('Mean F1 Weighted Score', fontsize=14)
plt.title('Mean F1 Weighted Score in Blocks of 10 Epochs', fontsize=16)
plt.xticks(x, [f'{i*10+1}-{(i+1)*10}' for i in x])
plt.legend(fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Add value labels on top of each bar
for i, values in enumerate(zip(original_aligned, warmup320_aligned, warmup100_aligned, HALF_warmup200_aligned)):
    positions = [i - 1.5*width, i - 0.5*width, i + 0.5*width, i + 1.5*width]
    
    for pos, val in zip(positions, values):
        if not pd.isna(val):  # Skip NaN values
            plt.text(pos, val + 0.01, f'{val:.3f}', 
                     ha='center', va='bottom', fontsize=8, rotation=90)

plt.tight_layout()
plt.savefig("comparison_f1_blocks.pdf")  # Save the plot as a PDF
plt.show()
```

```python
# Compute mean F1 scores in blocks of 10 epochs for all datasets
def mean_f1_by_blocks(df, epoch_col, f1_col, block_size=10):
    df_sorted = df.sort_values(epoch_col)
    blocks = (df_sorted[epoch_col] // block_size).astype(int)
    return df_sorted.groupby(blocks)[f1_col].mean()

# Calculate mean F1 by blocks for each dataset
datasets_blocks = {}
for name, df in datasets.items():
    filtered_df = df[df['epoch'] < 99]  # Ensure all datasets have similar range
    datasets_blocks[name] = mean_f1_by_blocks(filtered_df, 'epoch', 'eval_f1_weighted', block_size=10)

# Find all unique block indices across all datasets
all_blocks = set()
for block_means in datasets_blocks.values():
    all_blocks.update(block_means.index)
all_blocks = sorted(all_blocks)

# Align all series to the same index
aligned_series = {}
for name, block_means in datasets_blocks.items():
    aligned_series[name] = block_means.reindex(all_blocks)

# Plotting
plt.figure(figsize=(14, 8))
x = np.arange(len(all_blocks))
width = 0.14  # Narrower width to fit all bars
colors = ['tab:blue', 'tab:orange', 'tab:green', 'tab:red', 'tab:purple', 'tab:brown']

# Position offsets for bars
offsets = np.linspace(-0.4, 0.4, len(datasets))

# Create bars for each dataset
for i, (name, series) in enumerate(aligned_series.items()):
    plt.bar(x + offsets[i], series, width, label=name, color=colors[i])

plt.xlabel('Epoch Block', fontsize=12)
plt.ylabel('Mean F1 Weighted Score', fontsize=12)
plt.title('Comparison of Mean F1 Weighted Scores Across Models', fontsize=14)
plt.xticks(x, [f'{i*10+1}-{(i+1)*10}' for i in x])
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.legend(loc='lower right')
plt.tight_layout()

# Add value labels on top of each bar
for i, (name, series) in enumerate(aligned_series.items()):
    for j, value in enumerate(series):
        if not np.isnan(value):  # Skip NaN values
            plt.text(j + offsets[i], value + 0.01, f'{value:.2f}', 
                     ha='center', va='bottom', fontsize=8, rotation=90)

plt.savefig("model_comparison_f1_blocks.pdf")
plt.show()
```
