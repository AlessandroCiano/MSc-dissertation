```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer

df_userlevel = pd.read_csv(r'F:\VERO UTENTE\Desktop\Uni\dissertation\main\data\posts_with_labels.csv')
df_test = pd.read_csv(r'F:\VERO UTENTE\Desktop\Uni\dissertation\main\data\test_set.csv')

def ngram_match(post, userlevel_posts, n=7):
    # Create n-grams for the post
    try:
        vectorizer = CountVectorizer(ngram_range=(n, n), analyzer='word')
        ngrams_post = vectorizer.fit_transform([post])
        ngrams_post_set = set(vectorizer.get_feature_names_out())
        
        # For each userlevel post, check if any n-gram matches
        for ul_post in userlevel_posts:
            try:
                vectorizer_ul = CountVectorizer(ngram_range=(n, n), analyzer='word')
                ngrams_ul = vectorizer_ul.fit_transform([ul_post])
                ngrams_ul_set = set(vectorizer_ul.get_feature_names_out())
                if ngrams_post_set & ngrams_ul_set:
                    return True
            except ValueError:
                continue  # Skip posts that don't have enough words for n-grams
    except ValueError:
        return False  # Return False if the post doesn't have enough words for n-grams
    return False

userlevel_posts = df_userlevel['post'].tolist()
n = 7  # You can change n to the desired n-gram length

df_test['Duplicate'] = df_test['post'].apply(lambda x: ngram_match(x, userlevel_posts, n=n))
count_in_userlevel = df_test['Duplicate'].value_counts()
print(count_in_userlevel)

# print the rows in df_test that are unique (not in df_userlevel)
unique_posts = df_test[df_test['Duplicate'] == False]
print(unique_posts)

# Save the unique posts to a new CSV file
unique_posts.to_csv(r'F:\VERO UTENTE\Desktop\Uni\dissertation\main\data\test_set_30reduced.csv', index=False)
print(f"Saved {len(unique_posts)} unique posts to test_set_30reduced.csv")






```
