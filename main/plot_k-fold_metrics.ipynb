```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Load the data
cross_entropy_metrics_df = pd.read_csv('./K-fold/cross-entropy--5e-5\metrics/roberta-base_kfold_summary_metrics.csv')
ordinal_metrics_df = pd.read_csv('./K-fold/ordinal--5e-5/metrics/roberta-base_kfold_summary_metrics.csv')
soft_f1_entropy_metrics_df = pd.read_csv('./K-fold/soft_f1--5e-5/metrics/roberta-base_softf1_kfold_summary_metrics.csv')
```

```python
#cross_entropy_metrics metrics plot
cross_entropy_metrics_df = cross_entropy_metrics_df[cross_entropy_metrics_df['metric_type'] == 'averaged_per_epoch']


# Set the style
fig, ax1 = plt.subplots(figsize=(12, 6))

# Plot F1 weighted and Accuracy on the left y-axis
color = 'tab:red'
ax1.set_xlabel('Epoch', fontsize=14)
ax1.set_ylabel('F1 score', color=color, fontsize=14)
ax1.plot(cross_entropy_metrics_df['epoch'], cross_entropy_metrics_df['avg_eval_f1_weighted'], color=color, linestyle='-', marker='.', markersize=4, label='Validation F1 weighted')
#ax1.plot(cross_entropy_metrics_df['epoch'], cross_entropy_metrics_df['avg_eval_accuracy'], color='green', linestyle='-', marker='s', markersize=4, label='Validation Accuracy')
ax1.tick_params(axis='y', labelcolor=color)
ax1.grid(True)

# Create a second y-axis for the loss
ax2 = ax1.twinx()  # instantiate a second axes that shares the same x-axis

color = 'tab:blue'
ax2.set_ylabel('Loss', color=color, fontsize=14)  # we already handled the x-label with ax1
ax2.plot(cross_entropy_metrics_df['epoch'], cross_entropy_metrics_df['avg_eval_loss'], color=color, linestyle='-', marker='^', markersize=4, label='Validation Loss')
ax2.plot(cross_entropy_metrics_df['epoch'], cross_entropy_metrics_df['avg_train_loss'], color="#003C9C", linestyle='-', marker='^', markersize=4, label='Training Loss')
ax2.tick_params(axis='y', labelcolor=color)

# Add a title
plt.title('RoBERTa training: cross-entropy loss (avg. over 5 folds)', fontsize=16)

# Add legends
# To combine legends from both axes, we get handles and labels from both and pass them to a single legend command
lines, labels = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax2.legend(lines + lines2, labels + labels2, loc='best', fontsize=12)

fig.tight_layout()  # otherwise the right y-label is slightly clipped
plt.savefig("roberta-small_5fold_cross-entropy_training.pdf") # Save the plot as a PDF
plt.show()
```
